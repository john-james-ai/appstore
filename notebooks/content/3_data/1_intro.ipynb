## Background & Summary
From the core of the $6.3 billion app analytics market {cite}`marketsandmarketsAppAnalyticsMarket`, Apple’s App Store Ecosystem, comprised of some 40 software development kits (SDKs), Developer Academies, an Entrepreneur Camp, and over 250,000 application programming interfaces (APIs), delivers performance metrics, access to customer reviews, and a panoply of competitive and digital intelligence, allowing developers to retain customers, discover trends, cast product roadmaps and optimize return on investment {cite}`DevelopersGeneratedTrillion`. 

Further still, deep learning, natural language processing, large, 20 trillion word language models, and generative AI have redefined product/market strategy, and customer analytics; yet, the dearth of reliable, and publically available mobile customer experience benchmark datasets, constrains research and limits open, democratized, data-driven innovation. Without benchmark datasets, researchers, and data scientists are left following the available data, not the need. Benchmarking the mobile customer experience would serve four functions: 
1.	Enable the development of mobile-native and relevant algorithms for tasks such as product recommendation, language generation, and review summarization,
2.	Facilitate like-for-like comparisons of algorithmic performance, 
3.	Ensure reproducibility of algorithms, and
4.	Assert a normative influence on the way customer needs are discovered and served through innovation.

With emphasis, this is *not* a benchmark dataset! Rather, the VoCAI Dataset is an initial, work-in-progress, evolving set of Apple app, review, and rating data, collected and to be published periodically throughout 2023. The effort was motivated by three questions:
1.	What is the state, quality, and essence of the Apple mobile customer experience in the United States?
2.	How can state-of-the-art quantitative and qualitative methods expose unmet customer needs, facilitate entrepreneurial opportunity discovery, and value-based pricing optimization?
3.	How can generative AI enable prescriptive analytics and strategy formulation?

As of this writing, no large publically available mobile app rating and review datasets

## Rights and Permissions

Exploring the Apple customer experience As of May 2023, Apple's App Store offered over 1.76 million apps across 70 categories, accounting for over 62% of the global revenue share in 2022. This iterative, yet lean, exploratory data analysis (EDA) of over 300,000 apps across 26 categories aims to discover patterns, trends and clusters of opinion, and to generate actionable, high-impact insights for opportunity discover and analysis.

> “Far better an approximate answer to the right question, which is often vague, than an exact answer to the wrong question, which can always be made precise.” — John Tukey

The EDA is guided by the following nine, intentionally vague, yet provoking questions to stimulate discovery of the unknown, latent, yet high-impact/high-value, customer experience in an iterative, and likely chaotic learning process.

1. What best characterizes the nature of customer opinion within the mobile app market today?
2. What best characterizes the intensity of customer opinion?
3. How does opinion and customer satisfaction vary within and across mobile app categories?
4. Are their clusters exposed by the nature and intensity of opinion?
5. What does the distribution of rating and rating volume reveal about app creators / developers?
6. To what degree do these data suggest unmet customer needs, and fruitfull areas for opportunity discovery?
